# (1) Solving unconstrained optimization using gradient descent

```
# minimum value for stopping of gradient descent
epsilon=0.00001
```

```
# function for calculating the derivate
def derivate(f, X):
    h = 0.0000001
    delf = []
    
    for i in range(len(X)):
        E = np.zeros(len(X))
        E[i] = h
        vals = X + E
        delf.append((f(vals) - f(X))/h)
            
    return delf
```

```
# function for evaluation
def difference(X, Y):
    total = 0
    
    for i in range(len(X)):
        total = total + abs(X[i] - Y[i])
    total = total / len(X)
    

    return total
```

```
# function for implementing gradient descent
def steepest_descent(X, learning_rate,current_itr,iterations,function_val):
    
    while True:
        d = derivate(equation, X)
        x_prev = X
        
        X = X - np.dot(learning_rate, d)
        X = X.tolist()
        current_itr=current_itr+1
        f_val=equation(X)
        iterations.append(current_itr)
        function_val.append(f_val)
        if difference(x_prev, X) < epsilon:
            return x_prev
        
        
    return x_prev
```

```
# function for polynomial generation
def equation(x):
    
    #return 5*x[0]**2 - 30*x[0] + x[1]**2 -8*x[1] + 4*x[0]*x[1] + 34
    return (x[0]-2)**2 + (x[1]-4)**2 
```

```
# solution for (x-2)**2 + (y-4)**2
# initial points are taken as 5,6
iterations=[]
function_val=[]
learning_rate=float(input("Enter learning rate:"))
opt_points1=steepest_descent([5,6],learning_rate,0,iterations,function_val)

print("The optimal points for (x-2)**2 + (y-4)**2 are :", opt_points1)
print("The function value at optimal points is:",equation(opt_points1))
```

# (3) Solving constrained optimization problem

```
from sympy import *
```

## 3(a)Taking alpha=1

```
# taking x^2 + 4y^2 <=24 as active constraint
x, y, l1 = symbols('x y l1 ')

ans=solve([Eq(2*x*l1,0),
   Eq(3*y**2 + 8*y*l1 ,0),
   Eq(x**2 + 4*y**2 -24,0)], [x,y,l1])
print(ans)
# prints the list in the order of x,y,l1
```

```
# consider only positive lamda values
ans1=[x for x in ans if x[2]>0]
print(ans1)
```

```
# objective function
def objective(x):
  x1=x[0]
  x2=x[1]
  return x2**3
```

```
type(ans1)
```

```
#optimal variable values
final=ans1[0][:2]
print(final)
```

```
# final optimal function value
objective(final)
```

Therefore optimal function value for y^3 is -14.69.
Optimal points are 0,-2.44 for aplha=1

```
# taking both constraints as active 
x, y, l1,l2= symbols('x y l1 l2')

sol1=solve([Eq(2*x*l1+1*l2,0),
   Eq(3*y**2 + 8*y*l1 + l2,0),
   Eq(x**2 + 4*y**2-24,0),
   Eq(x+y,0)], [x,y,l1,l2])
print(sol1)
# prints the solution in the order of x,y,l1,l2.
```

```
# function for checking inactive constraints
for x in range(len(sol1)):
  if sol1[x][2]<0 or sol1[x][3]:
    print("lamda should not be negative")
```

Since lamda values are negative,this is not the optimal solution.

```
# considering x+y<=0 as active constraint
x, y,l2= symbols('x y l2')

solve([Eq(l2,0),
   Eq(3*y**2  + l2,0),
   Eq(x+y,0)], [x,y,l2])
```

Since in this case lamda is 0, consider as inactive constraint. Therefore, not the optimal solutiion.

##Another method for solving optimization problem with constraints using scipy

```
from scipy.optimize import minimize
```

```
def objective(x):
  x1=x[0]
  x2=x[1]
  return x2**3
```

```
def constarint1(x):
  return (x[0]**2 + 4*(x[1]**2) - 24)*(-1)
```

```
def constarint2(x):
  return (-1)*(x[0]+x[1])
```

```
def constarint3(x):
  return (-1)*(x[0]-x[1])
```

```
x0=[1,2]
print(objective(x0))
```

```
con1={'type': 'ineq','fun':constarint1}
con2={'type':'ineq','fun' : constarint2}
con3={'type':'ineq','fun' : constarint3}
cons=[con1,con2]
cons_=[con1,con3]
```

```
sol=minimize(objective,x0,method='SLSQP',constraints=cons)
```

```
# prints optimal function value,optimal points
print(sol)
```

Therefore optimal function value for y^3 is -14.69.
Optimal points are -1.71,-2.44 for aplha=1.

```
sol_=minimize(objective,x0,method='SLSQP',constraints=cons_,options={'disp':True})
```

```
# optimal points
xOpt=sol.x
print(xOpt)
```

Therefore optimal function value for y^3 is -10.51.
